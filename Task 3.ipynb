```python
import cv2
import numpy as np
import os

def enhance_v3(image_path):
    img = cv2.imread(image_path)
    if img is None: return
    
    name, ext = os.path.splitext(image_path)
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)
    
    mean = np.mean(l)
    
    if mean > 160:
        ratio = 120.0 / mean
        l = cv2.multiply(l, ratio)
        
        l = np.array(255 * (l / 255) ** 1.5, dtype='uint8')
        
    elif mean < 90:
        l = np.array(255 * (l / 255) ** 0.5, dtype='uint8')
        
    else:
        l = cv2.createCLAHE(clipLimit=3.0).apply(l)

    result = cv2.cvtColor(cv2.merge((l, a, b)), cv2.COLOR_LAB2BGR)
    cv2.imwrite(f"{name}_enhanced{ext}", result)

enhance_v3('bright.png')
```
