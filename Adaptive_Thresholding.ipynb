```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('dirty-img.png', cv2.IMREAD_GRAYSCALE)

# Apply Adaptive Thresholding
# Block Size (7): This determines the size of the pixel neighborhood. Larger = more global context. Smaller = captures fine details.
# C (23): It acts as a fine-tuning "buffer" to help filter out noise.

thresh_img = cv2.adaptiveThreshold(
    img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
    cv2.THRESH_BINARY, 7, 23
)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Grayscale")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Adaptive Gaussian Threshold")
plt.imshow(thresh_img, cmap='gray')
plt.axis('off')

plt.show()
```
